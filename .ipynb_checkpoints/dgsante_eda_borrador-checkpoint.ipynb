```python
df_mareas = pd.DataFrame.from_records(cursor.fetchall(), columns=[col[0] for col in cursor.description])
df_mareas['MAREA'] = pd.to_numeric(df_mareas['MAREA'])
cursor.close()
df_mareas.info()
```

```python
df_forms_dgsante = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vRsp2vmTU-CMxMpwWAui3SohhzlVSfQF-s5vGO3H8tkmFPjrKZk8ZZ47xflO-YQf4uMIkvVTiEX3aAe/pub?gid=0&single=true&output=csv')
df_forms_dgsante.info()
```

```python
df_forms_dgsante['1 -INICIO DE ANÁLISIS'] = pd.to_datetime(df_forms_dgsante['1 -INICIO DE ANÁLISIS']).dt.time
df_forms_dgsante['2 - FIN DE ANÁLISIS.1'] = pd.to_datetime(df_forms_dgsante['2 - FIN DE ANÁLISIS.1']).dt.time

df_forms_dgsante['INICIO_ANALISIS'] = df_forms_dgsante['1 - INICIO DE ANÁLISIS'] + ' ' +df_forms_dgsante['1 -INICIO DE ANÁLISIS'].astype(str)
df_forms_dgsante['FIN_ANALISIS'] = df_forms_dgsante['2 - FIN DE ANÁLISIS'] + ' ' +df_forms_dgsante['2 - FIN DE ANÁLISIS.1'].astype(str)
```

```python
df_forms_dgsante['INICIO_ANALISIS'] = pd.to_datetime(df_forms_dgsante['INICIO_ANALISIS'])
df_forms_dgsante['FIN_ANALISIS'] = pd.to_datetime(df_forms_dgsante['FIN_ANALISIS'])
```

```python
interesting_columns = ['MAREA','PLANTA','EMBARCACIÓN','INICIO_ANALISIS','FIN_ANALISIS']
df_forms_dgsante = df_forms_dgsante[interesting_columns]
df_forms_dgsante.head()
```

```python
df = pd.merge(df_mareas,df_forms_dgsante, on=["MAREA"],how='left')
df.info()
```

```python
df = df[df['TEMPORADA']!='']
dif_today = (pd.to_datetime(date.today()) - pd.to_datetime(inicio_tem_2022_2)).days  
date_previous_season = pd.to_datetime(inicio_tem_2022_1) + pd.DateOffset(days=dif_today)

print(date_previous_season)
print(pd.to_datetime(date.today()))
df.info()
```

```python
#df[(df['TEMPORADA']=='2022 - II') & (df['ACODERA'] < df['INICIO_ANALISIS'])][['ACODERA','INICIO_ANALISIS']]
df[~df['MAREA'].duplicated()]
# 
df['STATUS'] = np.where(df['ACODERA'] > df['INICIO_ANALISIS'], 0,1) 


#15 min dif -> inicio_Descarga hasta fin_analisis   --> 10min a mas 1 else 0 


#luego del analisis, una maximo para la descarga  10 min de desc hasta 1 hr +
# desde fin de analisis entre inicio de descarga ---> 15 como min 


# TABLA CHI
# D_INIC < h_arribo && D_INIC ES MADRUGADA ()--> fecha_de_descarga = fecha_de_arribo + 1

# ACODERAMIENTO > FECHA_DESCARGA --> STATUS = MALO

# FIN DE ANALISIS HASTA INICIO_SUCCION NO DEBE SER MAS DE 1:30 HORAS  --> INICIO_SUCCION = FECHA_DESCARGA

# P => ACODERAMIENTO HASTA INICIO_SUCCION



# porcentaje de datos erroneos --> acoderamiento > inicio_analisis


df['INICIO_ANALISIS'] = np.where(df['STATUS'] == 0, df['ACODERA'],df['INICIO_ANALISIS']) 
df.info()
```

```python
df['P'] = (df['INICIO_SUCCION'] - df['ACODERA'] ).dt.total_seconds() / 60.0
df['IAC'] = (df['INICIO_ANALISIS'] - df['ACODERA'] ).dt.total_seconds() / 60.0
df['IAN'] = (df['FIN_ANALISIS'] - df['INICIO_ANALISIS']).dt.total_seconds() / 60.0
df['FAN'] = (df['INICIO_SUCCION'] - df['FIN_ANALISIS']).dt.total_seconds() / 60.0
df['ISUC'] = (df['TERMINO_SUCCION'] - df['INICIO_SUCCION']).dt.total_seconds() / 60.0
df['FSUC'] = (df['DESACODERA'] - df['TERMINO_SUCCION']).dt.total_seconds() / 60.0
df.head()
```

```python
df2 = df.groupby(['TEMPORADA'],as_index=False).agg({'P':'mean','IAC':'mean','IAN':'mean','FAN':'mean','ISUC':'mean','FSUC':'mean'})
df2.rename(columns = {'P':'X̅ P','IAC':'X̅ IAC','IAN':'X̅ IAN','FAN':'X̅ FAN','ISUC':'X̅ ISUC','FSUC': 'X̅ FSUC'}, inplace = True)

df3 = df.groupby(['TEMPORADA'],as_index=False).agg({'P':'std','IAC':'std','IAN':'std','FAN':'std','ISUC':'std','FSUC':'std'})
df3.rename(columns = {'P':'σ P','IAC':'σ IAC','IAN':'σ IAN','FAN':'σ FAN','ISUC':'σ ISUC','FSUC': 'σ FSUC'}, inplace = True)

df4 = pd.merge(df2,df3,on=['TEMPORADA'],how='inner')
df4
```

```python
df[(df['TEMPORADA']=='2022 - II') & (df['ACODERA'] < df['INICIO_ANALISIS'])][['ACODERA','INICIO_ANALISIS']]
```

```python
df4['% IAC'] = df4['X̅ IAC']/df4['X̅ P']
df4['% IAN'] = df4['X̅ IAN']/df4['X̅ P']
df4['% FAN'] = df4['X̅ FAN']/df4['X̅ P']
df4['% ISUC'] = df4['X̅ ISUC']/df4['X̅ P']
df4['% FSUC'] = df4['X̅ FSUC']/df4['X̅ P']
df4
```

```python
df5 = df4[['TEMPORADA','X̅ P','X̅ IAC','X̅ IAN','X̅ FAN','X̅ ISUC','X̅ FSUC']]
df_medians = pd.melt(df5, id_vars='TEMPORADA', value_vars=['X̅ P', 'X̅ IAC', 'X̅ IAN','X̅ FAN','X̅ ISUC','X̅ FSUC'])
df_medians
```

```python
sns.lineplot(df_medians,x='variable',y='value',hue='TEMPORADA')
```

```python
df6 = df4[['TEMPORADA','% IAC','% IAN','% FAN','% ISUC','% FSUC']]
df_percentage= pd.melt(df6, id_vars='TEMPORADA', value_vars=['% IAC', '% IAN','% FAN','% ISUC','% FSUC'])
df_percentage['value']= df_percentage['value'].fillna(0)
df_percentage
```

```python
sns.barplot(df_percentage,x='variable',y='value',hue='TEMPORADA')
```
